Certainly there are outliers in our data. When we remove them, this leads to loss of information and moreover we have a very small dataset of 180 rows. So instead of removing, we do clipping the data between the (5 percentile and 95 percentile)

```python
box_ans = ["Miles", "Usage", "Age", "Income", "Fitness"]
for i in box_ans:
    percentile = AF[i].quantile([0.5,0.95]).values
    AF_COPY[i]= np.clip(AF_COPY[i],percentile[0],percentile[1])
```

Analysis of Categorical Columns with the Product

```python
analsis_col = ["Gender","MaritalStatus", "Usage","Fitness","Education", "Income_bins","Miles_bins","Age_bins"]
```

```python
for i in analsis_col:
    filter = AF[AF[i].notnull()]
    filter1 = pd.crosstab(filter[i], AF["Product"])
    row_sum = filter1.apply(sum ,axis = 1)
    ans = filter1.div(row_sum, axis = 0)
    final = round(ans,2)
    ax = plt.figure(figsize=(15,8))
    chat = final.plot(title = i, kind = "bar",width = 1.0)
    plt.show()
     
    chat.set_xlabel(i)
    chat.set_ylabel("Proportion")
    for contain in chat.containers:
        chat.bar_label(contain, fmt='%.1f', padding=3,rotation = 360)
        
    
```

MARGINAL PROBABILITY : Divide the row or column total by the total sample size. Marginal probability is the probability of an event occurring without any consideration of other events.

CONDITIONAL PROBABILITY : Conditional probability is a measure of the probability of an event occurring given that another event has already occurred. It is denoted as P(A∣B)

NORMALIZE : Normalize by dividing all values by the sum of values.
